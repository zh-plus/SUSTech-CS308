```python
from time import time

from matplotlib.pyplot import figure, imshow, show, subplot
import numpy as np
from PIL import Image

def imshow2(im1, im2=None):
    figure(figsize=[12, 15]) 
    if im2 is not None:
        subplot(1, 2, 1)
        imshow(np.round(im1 * 255).astype('uint8'), cmap='gray')
        subplot(1, 2, 2)
        imshow(np.round(im2 * 255).astype('uint8'), cmap='gray')
    else:
        imshow(np.round(im1 * 255).astype('uint8'), cmap='gray')  
    show()

pairs = [
    ('dog.bmp', 'cat.bmp'),
    ('einstein.bmp', 'marilyn.bmp'),
    ('fish.bmp', 'submarine.bmp'),
    ('bird.bmp', 'plane.bmp'),
    ('bicycle.bmp', 'motorcycle.bmp')
]
selected = 0
img1 = np.array(Image.open(pairs[selected][0]), dtype='float32')
img2 = np.array(Image.open(pairs[selected][1]), dtype='float32')
img1 /= 255
img2 /= 255

imshow2(img1, img2)
```

```python
def gaussian(krnl_size: tuple, std):
    krnl_h, krnl_w = krnl_size
    rh = krnl_h // 2
    rw = krnl_w // 2
    x, y = np.meshgrid(range(-rw, rw+1), range(-rh, rh+1))
    krnl = np.exp(-(x * x + y * y) / (2 * std * std))
    krnl[krnl < np.finfo('float32').eps * np.max(krnl)] = 0
    krnl /= np.sum(krnl)
    return krnl

krnl = gaussian((19, 19), 9)
figure()
imshow(krnl)
show()
```

```python
def conv2d_loop(image, filter):
    """
    Input
        image: a 3d array with shape of [H, W, 3]
        filter: a 2d array with shape of [fH, fW]
    Output
        a 3d array with shape of [H, W, 3]
    """
    h, w, channels = image.shape
    fh, fw = filter.shape[:2]
    pad_h = (fh - 1) // 2
    pad_w = (fw - 1) // 2
    image = np.pad(image, [(pad_h, pad_h), (pad_w, pad_w), (0, 0)], 'symmetric')

    filter = filter[..., None]  # [fH, fW] -> [fH, fW, 1]
    output = np.zeros((h, w, channels), dtype='float32')
    for r in range(h):
        for c in range(w):
            # image patch: [fH, fW, 3]
            # `filter`: [fH, fW, 1] -> [fH, fW, 3]
            result = image[r:r+fh, c:c+fw] * filter
            output[r, c, :] = np.sum(np.sum(result, axis=0), axis=0)
    return output


def conv2d(image, filter):
    """
    Input:
        image: a 3d array with shape of [H, W, 3]
        filter: a 2d array with shape of [fH, fW]
    Output:
        an array with shape [H, W, 3]
    """
    fh, fw = filter.shape[:2]
    pad_h = (fh - 1) // 2
    pad_w = (fw - 1) // 2
    image = np.pad(image, [(pad_h, pad_h), (pad_w, pad_w), (0, 0)], 'symmetric')

    h, w, c1 = image.shape
    new_shape = (h - fh + 1, w - fw + 1, fh, fw, c1)
    strides = np.array([w*c1, c1, w*c1, c1, 1]) * image.itemsize

    # image: [H, W, 3] -> [H, W, fH, fW, C]
    # filter: [fH, fW]
    image = np.lib.stride_tricks.as_strided(image, new_shape, strides)
    return np.tensordot(image, filter, axes=[(2, 3), (0, 1)])
```

```python
start = time()
img1_low = conv2d(img1, krnl)
print(f'Method 1 - {time() - start:.2f}/s')

start = time()
img1_low_2 = conv2d_loop(img1, krnl)
print(f'Method 2 (loop x 2) - {time() - start:.2f}/s')

imshow2(img1_low)
```

```python
#img2_high = conv2d(img2, gaussian((3, 3), 0.5)) - conv2d(img2, gaussian((11, 11), 5))
img2_high = img2 - conv2d(img2, gaussian((19, 19), 9))
img2_high = (img2_high - img2_high.min()) / np.ptp(img2_high)
imshow2(img2_high)
```

```python
hybird = img1_low + img2_high
hybird = (hybird - hybird.min()) / np.ptp(hybird)
figure(figsize=[8, 8])
imshow(hybird)

figure(figsize=[4, 4])
imshow(hybird)

figure(figsize=[2, 2])
imshow(hybird)

figure(figsize=[1, 1])
imshow(hybird)

show()
```
